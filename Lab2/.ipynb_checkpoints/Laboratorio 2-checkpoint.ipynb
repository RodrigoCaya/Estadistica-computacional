## 1. La paradoja del cumpleaños

La paradoja del cumpleaños es un problema matemático cuya respuesta es sorprendente dado el contexto del problema. En esta sección usted trabajará con este problema y podrá encontrar la respuesta que sorprendió a muchos. La paradoja es la siguiente: $N$ personas se reunen casualmente ¿Cuál es la probabilidad de que al menos 2 de ellas estén de cumpleaños el mismo día? **Considere año normal de 365 días.**

> a) Plantee la obtención de esta probabilidad en función de N. *Hint:* $\frac{Casos\;favorables}{Casos\;posibles}$


## Tenemos 2 sucesos.

### * Suceso A={"Al menos 2 personas celebran su cumpleaños a la vez"}
### * Suceso A^c= {"No hay 2 personas que celebren su cumpleaños a la vez"}

## Entonces sabemos que los Casos Posibles de un cumpleaños para n personas sera 365^n.

## Los Casos Favorables seran los dias que podria haber nacido en uno de los 365, luego en los 364 restantes... hasta 365-n.

## Finalmente, tendremos que los (casosFavorable)/(casosPosibles) seran igual a el suceso A^c.

## Para obtener el suceso A entonces 

# A=1-A^c

## Obteniendo el porcentaje de que ocurra A.



```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def probabilidadT(N):
    casosPosibles=365
    casosFavorables=365
    dias=365
    i=1
    while (i<N):
        casosPosibles=casosPosibles*dias
        casosFavorables=casosFavorables*(dias-i)
        i=i+1
    pNoA=casosFavorables/casosPosibles
    final=1-pNoA
    return final
print("Aqui se muestra la ecuacion aplicada en Python")
print("Por ejemplo la probabilidad de que 2 personas esten")
print("De cumpleaños el mismo dia en una reunion de 23 personas es:")
print(probabilidadT(23)*100,"%")


def mismoDia(lista, cantidad):
    if(cantidad==2 and len(lista)>=len(set(lista))+1):
        return 1
    elif(cantidad==3 and len(lista)>=len(set(lista))+2):
        return 1
    elif(cantidad==4 and len(lista)>=len(set(lista))+3):
        return 1
    elif(cantidad==5 and len(lista)>=len(set(lista))+4):
        return 1
    elif(cantidad==6 and len(lista)>=len(set(lista))+5):
        return 1
    else:
        return 0
def aleatorio(lista, cota):
    k=0
    final=[]
    while(k<cota):
        final.append(random.sample(lista,1)[0])
        k=k+1
    return final
```

## 2. Redes Complejas



El tema de Redes Complejas (grafos con muchos nodos) ha estado creciendo últimamente, sus orígenes se remontan a fines de los 90. Una de las propuestas para modelar la construcción de estas redes es la siguiente:

1. Se inicia con un solo nodo 
2. En cada iteración se agrega un nuevo nodo.
3. La probabilidad de unirse a un nodo viene dado por la siguiente expresión: $ P(Unirse\,al\,nodo_i) = \frac{Grado\,Nodo_i}{Grado\,Total\,Grafo}$ donde el Grado Total del Grafo corresponde a la suma de los grados de cada nodo perteneciente a este.
4. Se repite (2) y (3) varias veces hasta tener una red bastante grande.

A este tipo de red se les conoce como "Scale Free".

Un listado de links para refrescar conceptos: 

* [Grafo](https://es.wikipedia.org/wiki/Grafo)
* [Grado de un nodo](https://es.wikipedia.org/wiki/Grado_(teor%C3%ADa_de_grafos))


>a) Simule grafos de: 100, 1000 y 5000 nodos de la forma indicada y haga un scatterplot de Cantidad de Nodos v/s Grado del Nodo para los 3 grafos.

* Recuerde que las probabilidades se modifican después de agregar un nodo.
* Puede hacer una única iteración de 5000 pasos y guardar el estado de la red en 1000, 5000 y 10000.
* La matriz de adyacencia le facilitará el trabajo de obtener el grado de cada nodo (no es necesario hacerla, pero si tiene problemas encontrando alguna estructura de datos, esta le puede servir).

Matriz de Adyacencia:

Es una matriz cuadrada de $NxN$ donde en la posición $i,j$ hay un 1 si entre los nodos $i$ y $j$ hay una arista, de lo contrario esa entrada es 0. La diagonal en este casó será de 0 solamente.




```python
import numpy as np
import matplotlib.pyplot as plt
import random

def graficar(lista_grados,n):
    lista_grafico=[]
    maximo = max(lista_grados)
    for i in range(1,maximo+1):
        lista_grafico.append(lista_grados.count(i))
    plt.scatter(range(1,maximo+1),lista_grafico,color="red",edgecolors="black")
    plt.xlabel("Grados")
    plt.ylabel("Cantidad")
    plt.show()
    print ("Se realizaron pruebas con %d nodos."%(n))


def actualizarProb(lista_grados,lista_prob,n):
    for i in range(0,n-1):
        lista_prob[i] =  lista_grados[i]/((n-1)*2)
    lista_prob.append(1/((n-1)*2))
    return lista_prob



random.seed(96)
lista_nombres=[1,2]
lista_grados=[1,1]
lista_prob=[0.5,0.5]
L=1000    #primera prueba
M=5000    #segunda prueba
N=10000   #tercera prueba
for i in range(3,N+1):
        nodo_escogido = np.random.choice(lista_nombres, 1, lista_prob)
        lista_nombres.append(i)
        lista_grados.append(1)
        lista_grados[nodo_escogido[0]-1] += 1
        lista_prob = actualizarProb(lista_grados,lista_prob,i)
        if(i == L):
            graficar(lista_grados,L)
        if(i == M):
            graficar(lista_grados,M)
        
graficar(lista_grados,N)
   
    
```

>b) Observando los gráficos realizados en a) ¿Qué puede decirse acerca de la distribución de grado?

Debido a que la fórmula favorece la unión de nodos que tienen mayor grado, es normal que los datos se distribuyan de tal manera que exista una gran cantidad de nodos con grado uno (aproximadamente la mitad).
Su distribución se asemeja a una distribución de ley potencial, donde su probabilidad viene dado por el modelo P(K)=K^(-gamma), donde K es el grado del nodo.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from copy import copy, deepcopy


def quitarAristas(lista_grados,p,matriz):
    p*=100
    k=0
    for i in range(0,len(matriz)):
        if(matriz[i]==None):
            break
        else:
            for j in range(0,len(matriz[i])):
                try:
                    if(matriz[i][j]==None):
                        break
                except:
                    break
                else:
                    r = random.randint(0, 100)
                    ady = matriz[i][j]
                    matriz[i].remove(ady)
                    matriz[ady-1].remove(i+1)
                    if (r<=p):
                        lista_grados[i]-=1
                        lista_grados[ady-1]-=1
                        k+=1
    return (lista_grados,k)

def graficar(lista_grados,n,p,matriz):
    (lista_grados,k) = quitarAristas(lista_grados,p,matriz)
    lista_grafico=[]
    maximo = max(lista_grados)
    for i in range(1,maximo+1):
        lista_grafico.append(lista_grados.count(i))
    plt.scatter(range(1,maximo+1),lista_grafico,color="red",edgecolors="black")
    plt.xlabel("Grados del Nodo")
    plt.ylabel("Nodos")
    plt.show()
    print ("Se realizaron pruebas con p=%f para 10000 nodos."%(p))
    print ("Se eliminaron %d aristas."%(k))


def actualizarProb(lista_grados,lista_prob,n):
    for i in range(0,n-1):
        lista_prob[i] =  lista_grados[i]/((n-1)*2)
    lista_prob.append(1/((n-1)*2))
    return lista_prob


random.seed(1)
lista_nombres=[1,2]
lista_grados=[1,1]
lista_prob=[0.5,0.5]
matriz=[[2],[1]]
N=10000
p1=0.1
p2=0.4
p3=0.6
p4=0.9
for i in range(3,N+1):
        nodo_escogido = np.random.choice(lista_nombres, 1, lista_prob)
        lista_nombres.append(i)
        matriz[nodo_escogido[0]-1].append(i)
        matriz.append([nodo_escogido[0]])
        lista_grados.append(1)
        lista_grados[nodo_escogido[0]-1] += 1
        lista_prob = actualizarProb(lista_grados,lista_prob,i)

lista_grados_1=lista_grados.copy()
matriz_1=deepcopy(matriz)
graficar(lista_grados_1,N,p1,matriz_1)

lista_grados_2=lista_grados.copy()
matriz_2=deepcopy(matriz)
graficar(lista_grados_2,N,p2,matriz_2)

lista_grados_3=lista_grados.copy()
matriz_3=deepcopy(matriz)
graficar(lista_grados_3,N,p3,matriz_3)

lista_grados_4=lista_grados.copy()
matriz_4=deepcopy(matriz)
graficar(lista_grados_4,N,p4,matriz_4)   
    
```

Su distribución se mantiene pero con menos nodos y su forma sigue siendo la de libre escala, ya que los nodos que desaparecen son los que tienen más grados (dependiendo que tan grande es el "p") y esto hace que la curva sea más corta, pero su forma se mantiene.

# EVOLUCIÓN PUNTAJE 1980

```python
import matplotlib.pyplot as plt

lista_promedio_1 = []
mayor = finaldf.loc[finaldf['year_of_birth'] >= 1980]
lista_promedio_1.append(mayor.loc[mayor['year_of_pub'] == 2005][['promedio']].mean())
lista_promedio_1.append(mayor.loc[mayor['year_of_pub'] == 2010][['promedio']].mean())
lista_promedio_1.append(mayor.loc[mayor['year_of_pub'] == 2015][['promedio']].mean())
lista_promedio_1.append(mayor.loc[mayor['year_of_pub'] == 2019][['promedio']].mean())

lista_promedio_2 = []
menor = finaldf.loc[finaldf['year_of_birth'] < 1980]
lista_promedio_2.append(menor.loc[menor['year_of_pub'] == 2005][['promedio']].mean())
lista_promedio_2.append(menor.loc[menor['year_of_pub'] == 2010][['promedio']].mean())
lista_promedio_2.append(menor.loc[menor['year_of_pub'] == 2015][['promedio']].mean())
lista_promedio_2.append(menor.loc[menor['year_of_pub'] == 2019][['promedio']].mean())

plt.plot([2005,2010,2015,2019],lista_promedio_1,label="mayor 1980",color="blue")
plt.plot([2005,2010,2015,2019],lista_promedio_2,label="menor 1980",color="red")
plt.legend()

plt.ylabel('Promedio')
plt.xlabel('Años')
plt.title("Evolución puntaje")


plt.show()
```

Se puede evidenciar que en el año 2015 los artistas que nacieron despues de 1980 tuvieron un "peak" de puntaje, y a su vez los artistas que nacieron antes de 1980 disminuyeron su puntaje. Esto quiere decir que durante ese periodo la música más "nueva" tuvo mejor rating que la musica "antigua" (considerando que la gente joven hace musica nueva y la vieja no).

# LINEA DE TIEMPO

```python
import collections
lista_por_anios = []
for i in finaldf.year_of_pub.unique():
    groups_by_anio = finaldf.loc[finaldf.year_of_pub == i] 
    groups_by_anio = groups_by_anio.groupby(['year_of_pub','genre']).count().sort_values('album_title',ascending=False).head(3)
    lista_por_anios.append(list(groups_by_anio.index.values))

dic = {}
lista_generos = []
for i in range(0,len(lista_por_anios)):
    (_,gen_1)=lista_por_anios[i][0]
    (_,gen_2)=lista_por_anios[i][1]
    (_,gen_3)=lista_por_anios[i][2]
    lista_generos.append([gen_1,gen_2,gen_3])
    dic[lista_por_anios[i][0][0]] = lista_generos[i]

dic_2 = collections.OrderedDict(sorted(dic.items()))

pd.DataFrame(dic_2)
```

Se evidencia la tendencia de que todos son iguales, no hay cambio en ningun año. En todos es Indie, Pop y Rap.
